# 1.1) Fonction pour obtenir et analyser le contenu html d'une page Wikipedia :

```python
import requests
from bs4 import BeautifulSoup

def get_wikipedia_page(url):
    # Récupère le contenu html de la page
    response = requests.get(url)
    html = response.content
    # Parse le html avec BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

# 1.2) Fonction pour extraire le titre de l'article :

```python
def get_article_title(soup):
    title = soup.find('title')
    return title.text
```

# 1.3) Fonction pour extraire le texte de l'article pour chaque paragraphe avec leurs rubriques. Associez ces titres à leurs paragraphes respectifs dans le dictionnaire :

```python
def get_article_content(soup):
    # Récupère tous les éléments de type "h2" et "p" dans l'article
    headers = soup.find_all('h2')
    paragraphs = soup.find_all('p')

    # Initialise le dictionnaire de contenu
    content = {}

    # Ajoute le texte de chaque paragraphe à son titre correspondant dans le dictionnaire
    for header in headers:
        header_text = header.text.strip()
        content[header_text] = []
        for paragraph in paragraphs:
            if paragraph.previous_sibling == header:
                content[header_text].append(paragraph.text.strip())

    return content
```

# 1.4) Fonction pour collecter chaque lien qui redirige vers une autre page Wikipédia :



```python
def get_wikipedia_links(soup):
    # Récupère tous les liens de la page
    links = soup.find_all('a')

    # Initialise la liste des liens wikipédia
    wikipedia_links = []

    # Ajoute les liens pointant vers une page wikipédia à la liste
    for link in links:
        href = link.get('href')
        if href and href.startswith('/wiki/'):
            wikipedia_links.append('https://fr.wikipedia.org' + href)

    return wikipedia_links
```

# 1.5) Fonction pour englober toutes les fonctions précédentes dans une seule fonction qui prend comme paramètre un lien Wikipedia :

```python
def analyze_wikipedia_page(url):
    soup = get_wikipedia_page(url)
    title = get_article_title(soup)
    content = get_article_content(soup)
    links = get_wikipedia_links(soup)

    wikipedia_data = {
        'title': title,
        'content': content,
        'links': links
    }

    return wikipedia_data
```

# 1.6) Testez la dernière fonction sur une page Wikipédia de votre choix :

```python
url = 'https://fr.wikipedia.org/wiki/Python_(langage)'
data = analyze_wikipedia_page(url)
print(data['title'])
print(data['content'])
print(data['links'])
```
